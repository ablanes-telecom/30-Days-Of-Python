Modifying the data frame

```python
#Creating a DataFrame
data = [
    {"Name": "Asabeneh", "Country":"Finland","City":"Helsinki"},
    {"Name": "David", "Country":"UK","City":"London"},
    {"Name": "John", "Country":"Sweden","City":"Stockholm"}]
df = pd.DataFrame(data)
print(df)
```

```python
#adding a column 
weights = [74, 78, 69]
df['Weight'] = weights
heights = [173, 175, 169]
df['Height'] = heights
print(df)
```

```python
#Modifing columns values
df['Height'] = df['Height'] *0.0001
print(df)
```

```python
#ahora creamos una columna que nos calcula el peso por la altura:
# Using functions makes our code clean, but you can calculate the bmi without one
def calculate_bmi ():
    weights = df['Weight']
    heights = df['Height']
    bmi = []
    for w,h in zip(weights, heights):
        b = w/(h*h)
        bmi.append(b)
    return bmi

bmi = calculate_bmi()
df['BMI'] = bmi
df
```

```python
#Formatting DataFrame columns
df['BMI'] = round(df['BMI'], 1)
print(df)
```

```python
#añadimos dos columnas mas
birth_year = ['1769', '1985','1990']
current_year = pd.Series(2020, index=[0, 1,2])
df['Birth Year'] = birth_year
df['Current Year'] = current_year
df
```

```python
df['Birth Year'].dtype # it gives string object , we should change this to number
df['Birth Year'] = df['Birth Year'].astype('int')
print(df['Birth Year'].dtype) # let's check the data type now
```

```python
df['Current Year'] = df['Current Year'].astype('int')
df['Current Year'].dtype
```

```python
ages = df['Current Year'] - df['Birth Year'] #ahora ya podemos hacer operaciones con ellos
ages
```

```python
df['Ages'] = ages
print(df)
```

```python
#hay un dato que esta mal así que para cambiarlo hacemos la mediana del resto
mean = int((35 + 30)/ 2)
print('Mean: ',mean)	#it is good to add some description to the output, so we know what is what
df.loc[0, 'Ages'] = mean
print(df)
```

```python
df['Ages']=df['Ages'].astype(int)
print(df)
```
